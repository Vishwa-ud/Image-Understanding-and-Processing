# Histogram calculation/plotting in OpenCV

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread("D:/UNI_FILEs/YEAR_4/Y4S2/IUP/Week 3 - 20 February - 26 February/Lab_03/sunflower.jpg", 0)

# Find frequency of pixels in range 0 - 255 - histogram
hist = cv2.calcHist([img], [0], None, [256], [0, 256])

# Show the plotting graph of an image
plt.plot(hist)
plt.show()
```

# Histogram calculation/plotting in Numpy

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread("D:/UNI_FILEs/YEAR_4/Y4S2/IUP/Week 3 - 20 February - 26 February/Lab_03/sunflower.jpg", 0)

# Find frequency of pixels in range 0 - 255 - Histogram
hist, bins = np.histogram(img.ravel(), 256, [0, 256])

# Show the plotting graph of an image
plt.plot(hist)
plt.show()
```

# BGR histograms – Complete the code below

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread("D:/UNI_FILEs/YEAR_4/Y4S2/IUP/Week 3 - 20 February - 26 February/Lab_03/sunflower.jpg")
color = ('b', 'g', 'r')

for i,col in enumerate(color):
    histr = cv2.calcHist([img], [i], None, [256], [0,256])
    plt.plot(histr, color = col)
    plt.xlim([0,256])
plt.show()
```

```python
img = cv.imread("sunflower.jpg", cv.IMREAD_GRAYSCALE)

# create a mask
mask = np.zeros(img.shape[:2], np.uint8)

mask[150:600, 500:1100] = 255

masked_img = cv.bitwise_and(img, img, mask = mask)

# Calculate histogram with mask and without mask
# Check third argument for mask
hist_full = cv.calcHist([img], [0], None, [256], [0,256])
hist_mask = cv.calcHist([img], [0], mask, [256], [0,256])

plt.subplot(221), plt.imshow(img, 'gray')
plt.subplot(222), plt.imshow(mask,'gray')
plt.subplot(223), plt.imshow(masked_img, 'gray')
plt.subplot(224), plt.plot(hist_full), plt.plot(hist_mask)
plt.xlim([0,256])

plt.show()
```
