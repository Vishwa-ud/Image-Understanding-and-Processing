### 1. Histogram calculation/plotting in OpenCV 

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('D:\\Education\\IUP\\Lab3\\sunflower.jpg',1)
```

```python
color = ('b','g','r')
for i, col in enumerate(color):
    hist1 = cv2.calcHist([img], [i], None, [256], [0,256])
    plt.plot(hist1, color = col)
    plt.show
```

### 2. Histogram calculation/plotting in Numpy 

```python
b, g, r = cv2.split(img)
```

```python
hist2, bins = np.histogram(b.ravel(), 256, [0,256])
hist3, bins = np.histogram(g.ravel(), 256, [0,256])
hist4, bins = np.histogram(r.ravel(), 256, [0,256])
```

```python
 plt.plot(hist2)
 plt.plot(hist3)
 plt.plot(hist4)

plt.show()
```

### 3. BGR histograms – Complete the code below

```python
mask = np.zeros(img.shape[0:2], np.uint8)#create mask full black mask complete black img
mask [50:300, 250:400] = 255
plt.imshow(cv2.cvtColor(mask, cv2.COLOR_BGR2RGB))
```

```python
hist5 = cv2.calcHist([mask], [0], None, [256], [0,256])#indencity brightness
plt.plot(hist5)
plt.show()
```

```python
maskedImage = cv2.bitwise_and(img, img, mask = mask)
plt.imshow(cv2.cvtColor(maskedImage,cv2.COLOR_BGR2RGB))
```

```python
color = ('b','g','r')

for i, col in enumerate(color):
    hist6 = cv2.calcHist([img], [i], mask, [256], [0,256])
    plt.plot(hist6, color = col)
    plt.show()
```
