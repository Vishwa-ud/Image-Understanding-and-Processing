## Color Histogram Equalization to improve the contrast of color image

```python
import cv2
import numpy as np
import matplotlib. pyplot as plt
image = cv2.imread('D:\\Education\\IUP\\Lab4\\color.png' , 1)

# convert image from RGB to HSV
#comptete the code here
img_hsv = cv2.cvtColor(image, cv2.COLOR_RGB2HSV)


# Histogram equalisation on the V-channel
img_hsv[:, :, 2] = cv2.equalizeHist(img_hsv[:, :, 2])

# convert image back from HSV to RGB
#compLete the code here
img_RGB = cv2.cvtColor(img_hsv, cv2.COLOR_HSV2RGB)

#PIot the two histograns and display the two images
#comptete the code here

histogram1 = cv2.calcHist([image], [0], None, [256], [0,256])
histogram2 = cv2.calcHist([img_RGB], [0], None, [256], [0,256])

plt.subplot(121), plt.plot(histogram1)
plt.subplot(122), plt.plot(histogram2)

results = np.hstack((image, img_RGB))
cv2. imshow( 'Results',results)
cv2. waitKey(0)
```
