```python
import cv2 as cv
import matplotlib.pyplot as plt

# Load an image from file
image = cv.imread('image.jpg')

# Check if the image is loaded successfully
if image is None:
    print("Error: Could not read the image file.")
else:
    # Convert BGR to RGB (Matplotlib uses RGB)
    image_rgb = cv.cvtColor(image, cv.COLOR_BGR2RGB)

    # Display the image using Matplotlib
    plt.imshow(image_rgb)
    plt.axis('off')  # Hide axis labels
    plt.title('Original Image')
    plt.show()
```

```python
# Convert BGR image to Grayscale
gray_image = cv.cvtColor(image, cv.COLOR_BGR2GRAY)

# Display the grayscale image using Matplotlib
plt.imshow(gray_image, cmap='gray')  # Use 'gray' colormap
plt.axis('off')  # Hide axis labels
plt.title('Grayscale Image')
plt.show()
```

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = cv.imread('image.jpg')

# Convert BGR image to HSV
hsv_image = cv.cvtColor(image, cv.COLOR_BGR2HSV)

# Define the range for YELLOW color in HSV
lower_yellow = np.array([20, 100, 100])  # Lower bound of yellow
upper_yellow = np.array([35, 255, 255])  # Upper bound of yellow

# Create a mask: Only yellow areas will be white, everything else will be black
mask = cv.inRange(hsv_image, lower_yellow, upper_yellow)

# Extract only the yellow object by applying the mask
yellow_object = cv.bitwise_and(image, image, mask=mask)

# Convert images to RGB for correct Matplotlib display
image_rgb = cv.cvtColor(image, cv.COLOR_BGR2RGB)
yellow_object_rgb = cv.cvtColor(yellow_object, cv.COLOR_BGR2RGB)

# Display results
fig, ax = plt.subplots(1, 3, figsize=(15, 5))

ax[0].imshow(image_rgb)
ax[0].set_title('Original Image')
ax[0].axis('off')

ax[1].imshow(mask, cmap='gray')
ax[1].set_title('Yellow Mask')
ax[1].axis('off')

ax[2].imshow(yellow_object_rgb)
ax[2].set_title('Extracted Yellow Object')
ax[2].axis('off')

plt.show()
```

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = cv.imread('image.jpg')

# Convert BGR image to HSV
hsv_image = cv.cvtColor(image, cv.COLOR_BGR2HSV)

# Define the range for YELLOW color in HSV
lower_yellow = np.array([20, 100, 100])  # Lower bound of yellow
upper_yellow = np.array([35, 255, 255])  # Upper bound of yellow

# Define the range for GREEN color in HSV
lower_green = np.array([35, 50, 50])  # Lower bound of green
upper_green = np.array([85, 255, 255])  # Upper bound of green

# Create masks
yellow_mask = cv.inRange(hsv_image, lower_yellow, upper_yellow)  # Mask for yellow
green_mask = cv.inRange(hsv_image, lower_green, upper_green)  # Mask for green

# Extract only the yellow flower and green leaves
yellow_object = cv.bitwise_and(image, image, mask=yellow_mask)
green_object = cv.bitwise_and(image, image, mask=green_mask)

# Convert images to RGB for Matplotlib display
image_rgb = cv.cvtColor(image, cv.COLOR_BGR2RGB)
yellow_object_rgb = cv.cvtColor(yellow_object, cv.COLOR_BGR2RGB)
green_object_rgb = cv.cvtColor(green_object, cv.COLOR_BGR2RGB)

# Display results
fig, ax = plt.subplots(1, 4, figsize=(20, 5))

ax[0].imshow(image_rgb)
ax[0].set_title('Original Image')
ax[0].axis('off')

ax[1].imshow(yellow_mask, cmap='gray')
ax[1].set_title('Yellow Mask')
ax[1].axis('off')

ax[2].imshow(green_mask, cmap='gray')
ax[2].set_title('Green Mask')
ax[2].axis('off')

ax[3].imshow(green_object_rgb)
ax[3].set_title('Extracted Green Leaves')
ax[3].axis('off')

plt.show()
```
