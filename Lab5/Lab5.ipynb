## 1. Negative transformation for color images

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('apples.jpg' , 1)

img_neg = cv2.bitwise_not(img)

plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.show()
plt.imshow(cv2.cvtColor(img_neg,cv2.COLOR_BGR2RGB))
plt.show()

#cv2.waitKey(O)
```

## 3. Power-law transformation to improve the contrast of  the following image

```python
img1 = cv2.imread('gamma.jpg' , 0)

#original image
plt.imshow(img1,cmap='gray')
plt.show()

#bright transformation
gamma_b = 0.5 # can select any valuse less than 1
img_power_b = np.power(img1,gamma_b)

plt.imshow(img_power_b,cmap='gray')
plt.show()

#dark transformation
gamma_d = 1.4 # can select any valuse grater than 1
img_power_d = np.power(img1,gamma_d)

plt.imshow(img_power_d,cmap='gray')
plt.show()
```

## 4. Log transformation to improve the dynamic range of an image

```python
img3 = cv2.imread('log.jpg' , 0)

np.seterr(divide="ignore")
np.seterr(invalid="ignore")

c = 255/np.log(1+np.max(img3))

logimg = c * (np.log(1+img3))

logimg = np.array(logimg,dtype=np.uint8)

plt.imshow(img3,cmap='gray')
plt.show() 

plt.imshow(logimg,cmap='gray')
plt.show()
```

```python
img = cv2.imread("log.jpg", 0)

# Apply log transformation -----------------------------------------------

# 1st Method
# c = 255 / np.log(1 + np.max(img))
# log_img = c * np.log( img + 1)

# 2nd Method
log_img = (np.log(img + 1 ) / (np.log( 1 + np.max( img )))) * 255

# Ignore the error
np.seterr(divide='ignore')
np.seterr(invalid="ignore")

# --------------------------------------------------------------------------

log_img = np.array(log_img, dtype= np.uint8)

plt.imshow(img, cmap="gray")
plt.show()

plt.imshow(log_img, cmap="gray")
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read image
img = cv2.imread("log.jpg", 0)  # Read as grayscale
img = img.astype(np.float64)  # Convert image to float64 to prevent overflow

# Apply log transformation method
np.seterr(divide='ignore', invalid='ignore')  # Ignore warnings

c = 255 / np.log(1 + np.max(img))  # Compute scaling factor
loging = c * (np.log(1 + img))  # Apply log transformation

# Normalize and convert back to uint8
log_image = np.clip(loging, 0, 255).astype(np.uint8)

# Display original image
plt.imshow(img, cmap='gray')
plt.title("Original Image")
plt.show()

# Display log-transformed image
plt.imshow(log_image, cmap='gray')
plt.title("Log Transformed Image")
plt.show()
```
