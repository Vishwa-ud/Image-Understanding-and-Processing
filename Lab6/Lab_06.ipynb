```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

# Image filtering using 2D convolution


```python
# Load the image
noise_img = cv.imread("D:/UNI_FILEs/YEAR_4/Y4S2/IUP/Week 5 - 20 March - 26 March/Lab_06/coins.png", 1)

# Apply custom averaging filter (Lowpass Filter) using filter2D()
kernal = np.ones((5, 5), np.float32)/25
dst = cv.filter2D(noise_img, -1, kernal)

result = np.hstack((noise_img, dst))
# cv.imshow("Filtered Image", result)
# cv.waitKey();
plt.imshow(cv.cvtColor(result, cv.COLOR_BGR2RGB))
plt.show()
```

# Image Averaging using Box Filter


```python
# Load the image
noise_img = cv.imread("D:/UNI_FILEs/YEAR_4/Y4S2/IUP/Week 5 - 20 March - 26 March/Lab_06/coins.png", 1)

# Apply averaging filter (Lowpass Filter) : (3 * 3 filter)
dst = cv.blur(noise_img, (3, 3))

result = np.hstack((noise_img, dst))
# cv.imshow("Filtered Image", result)
# cv.waitKey();
plt.imshow(cv.cvtColor(result, cv.COLOR_BGR2RGB))
plt.show()
```

```python
# Load the image
noise_img = cv.imread("D:/UNI_FILEs/YEAR_4/Y4S2/IUP/Week 5 - 20 March - 26 March/Lab_06/coins.png", 1)

# Apply averaging filter (Lowpass Filter) : (3 * 3 box filter)
dst = cv.boxFilter(noise_img, -1, (3, 3))

result = np.hstack((noise_img, dst))
# cv.imshow("Filtered Image", result)
# cv.waitKey();
plt.imshow(cv.cvtColor(result, cv.COLOR_BGR2RGB))
plt.show()
```

# Median Filtering and Gaussian Filtering


```python
# Load the image
noise_img = cv.imread("D:/UNI_FILEs/YEAR_4/Y4S2/IUP/Week 5 - 20 March - 26 March/Lab_06/saltandpepper1.png", 1)

# Apply averaging filter : (Median value = 3)
dst = cv.medianBlur(noise_img, 3)

result = np.hstack((noise_img, dst))
# cv.imshow("Filtered Image", result)
# cv.waitKey();
plt.imshow(cv.cvtColor(result, cv.COLOR_BGR2RGB))
plt.show()
```

```python
# Load the image
noise_img = cv.imread("D:/UNI_FILEs/YEAR_4/Y4S2/IUP/Week 5 - 20 March - 26 March/Lab_06/gaussian Noise.png", 1)

# Apply averaging filter : (Median value = 3)
dst = cv.GaussianBlur(noise_img, (11, 11), 0)

result = np.hstack((noise_img, dst))
# cv.imshow("Filtered Image", result)
# cv.waitKey();
plt.imshow(cv.cvtColor(result, cv.COLOR_BGR2RGB))
plt.show()
```

# Application of image blurring


```python
# Load the image
noise_img = cv.imread("D:/UNI_FILEs/YEAR_4/Y4S2/IUP/Week 5 - 20 March - 26 March/Lab_06/hubble.png", 0)

# Apply custom averaging filter (Lowpass Filter) using filter2D()
kernal = np.ones((15, 15), np.float32)/225
dst = cv.filter2D(noise_img, -1, kernal)

# Apply thresholding operator to highlight largest object
ret, thresh = cv.threshold(dst,110,255,cv.THRESH_BINARY)

# result = np.hstack((noise_img, dst, thresh_img))
result = np.hstack((noise_img, dst, thresh))

# cv.imshow("Filtered Image", result)
# cv.waitKey();
plt.imshow(cv.cvtColor(result, cv.COLOR_BGR2RGB))
plt.show()
```
