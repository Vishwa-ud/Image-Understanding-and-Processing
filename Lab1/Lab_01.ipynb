```python
img = cv.imread('D:/UNI_FILEs/YEAR_4/Y4S2/IUP/Week 1 - 6 February - 12 February/Lab_01/flower.jpg', 1)
cv.imshow('Original Image', img)

imgGray = cv.imread('D:/UNI_FILEs/YEAR_4/Y4S2/IUP/Week 1 - 6 February - 12 February/Lab_01/flower.jpg', 0)
cv.imshow('Gray Image', imgGray)

res, binary = cv.threshold(imgGray, 120, 255, cv.THRESH_BINARY)
cv.imshow('Binary Image', binary)

print(res)

cv.waitKey()
cv.destroyAllWindows()
```
